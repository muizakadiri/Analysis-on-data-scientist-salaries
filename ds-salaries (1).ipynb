```python
#visualizing multiple plots/subplots
fig, axes = plt.subplots(2,4, figsize=(20,10))
sns.countplot(x='work_year', data=data_science, ax=axes[0,0])
sns.countplot(x='company_location', data=data_science, ax=axes[0,1])
sns.countplot(x='employment_type', data=data_science, ax=axes[0,2])
sns.countplot(x='experience_level', data=data_science, ax=axes[0,3])
sns.countplot(x='remote_ratio', data=data_science, ax=axes[1,0])
sns.countplot(x='company_size', data=data_science, ax=axes[1,1])
sns.histplot(x='job_title', data=data_science, ax=axes[1,2])
sns.histplot(x='salary_in_usd', data=data_science, ax=axes[1,3])
```

from the above diagrams, 
1.work_year: there is a progression in the numbers of workers in the industry as there is a clear increase from year 2020 to 2021 and also 2021 to 2022 which signifies growth.
2.Employment_type:The contract of employment between the employer and the employee is more of a full time in the industry. probably due to the workload demmand
3. Experience_level: there are more people in the middle and senior level which means the new entrants have enough mentors to learn from.
4. remote_ratio: The number of the experienced workers exceed the new workers which means they require less supervision which translate to getting the job done with little or no supervision.
5.company_size: there are more medium(between 50 to 250 workers) size company operating in the industry

```python

plt.figure(figsize=(9,8))
plt.title('Graphic representation of salary in usd')
sns.countplot(x='job_title',data= data_science)
plt.xlabel('salary_in_usd')
plt.ylabel('count ')
plt.show
```

```python
#creating a visual to show the salary_in_usd
plt.figure(figsize=(9,8))
plt.title('Graphic representation of salary in usd')
plt.hist(x='job_title',data= data_science)
plt.xlabel('job_title')
plt.ylabel('count ')
plt.show
```

```python
#create a pivot bywork_year and salary_in_usd
work_salary_pivot = pd.crosstab(data_science.work_year,data_science.salary_in_usd,normalize='index')
work_salary_pivot.plot.bar(stacked=True)
plt.title('work_year analysis with salary base on year')
plt.show()
```

```python
work_salary_pivot.info()
```

```python
#create a pivot bywork_year and salary_in_usd
work_remote_pivot = pd.crosstab(data_science.work_year,data_science.remote_ratio,normalize='index')
work_remote_pivot.plot.bar(stacked=True)
plt.title('work_year analysis with remote_ratio')
plt.show()
```
